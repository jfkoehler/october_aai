```python
from skimage import io
img = io.imread('im_data/cats/cat.1003.jpg', as_gray=True)
```

```python
plt.imshow(img)
```

```python
newim = np.zeros(img.shape)
conv_filter = np.random.normal(size = 9).reshape(3, 3)
```

```python
for i in range(img.shape[0]-3):
    for j in range(img.shape[1] - 3):
        newim[i:i+3, j:j+3] = img[i:i+3, j:j+3]*conv_filter
        
```

```python
newim[:2, :2]
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 6))
ax[0].imshow(im, cmap = 'gray')
ax[1].imshow(newim, cmap = 'gray')
```
