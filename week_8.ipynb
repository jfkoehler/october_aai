```python
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

### A Basic Feed-Forward Network

```python
network_1 = Sequential()
network_1.add(Flatten())
network_1.add(Dense(32, activation = 'relu'))
network_1.add(Dense(10, activation = 'softmax'))

network_1.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                 metrics = ['acc'])
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_images = train_images/255
test_images  = test_images/255
```

```python
history = network_1.fit(train_images, train_labels, validation_data=(test_images, test_labels),
                       epochs = 20)
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
```

```python
network_2 = Sequential()
network_2.add(Flatten())
network_2.add(Dense(128, activation = 'relu'))
network_2.add(Dense(64, activation = 'relu'))
network_2.add(Dense(32, activation = 'relu'))
network_2.add(Dense(10, activation = 'softmax'))

network_2.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                 metrics = ['acc'])

history = network_2.fit(train_images, train_labels, validation_data=(test_images, test_labels),
                       epochs = 20)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
```

```python
from keras.layers import Dropout
from keras.reg
```

```python
network_3 = Sequential()
network_3.add(Flatten())
network_3.add(Dense(128, activation = 'relu'))
network_3.add(Dropout(0.3))
network_3.add(Dense(64, activation = 'relu'))
network_3.add(Dropout(0.4))
network_3.add(Dense(32, activation = 'relu'))
network_3.add(Dense(10, activation = 'softmax'))

network_3.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                 metrics = ['acc'])

history = network_3.fit(train_images, train_labels, validation_data=(test_images, test_labels),
                       epochs = 20)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
```

### Giving CNN a Shot

```python
train_images = train_images.reshape(train_images.shape[0], 28,28, 1)
test_images = test_images.reshape(test_images.shape[0], 28, 28, 1)
```

```python
cnnmodel = Sequential()

cnnmodel.add(Conv2D(filters = 6, kernel_size = (3, 3), activation = 'relu',
                   input_shape = (28,28, 1)))
cnnmodel.add(MaxPool2D(pool_size = (2, 2)))
cnnmodel.add(Conv2D(filters = 6, kernel_size = (3, 3), activation = 'relu'))
cnnmodel.add(MaxPool2D(pool_size = (2, 2)))


cnnmodel.add(Flatten())
cnnmodel.add(Dense(64, activation = 'relu'))
cnnmodel.add(Dense(128, activation = 'relu'))
cnnmodel.add(Dense(10, activation = 'softmax'))

cnnmodel.compile(loss = 'categorical_crossentropy', metrics = ['acc'])
history = cnnmodel.fit(train_images, train_labels, epochs = 20)
```

```python
plt.plot(history.history['acc'])
```

### Loading Custom Images and Training

```python
from keras.preprocessing import image_dataset_from_directory
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rotation_range=30,
                                   width_shift_range=.1,
                                   zoom_range=0.2,
                                   rescale = 1/255)
```

```python
train_generator = train_datagen.flow_from_directory('im_data/',
                                                   target_size = (100, 100))
```

```python
newmodel = Sequential()
newmodel.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (100, 100, 3)))
newmodel.add(MaxPool2D(2, 2))
newmodel.add(Flatten())
newmodel.add(Dense(32, activation = 'relu'))
newmodel.add(Dense(1, activation = 'sigmoid'))

newmodel.compile(loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
newmodel.fit(train_generator, epochs = 10)
```
