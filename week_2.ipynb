````python
from sklearn.datasets import load_digits
````

````python
digits = load_digits()
````

````python
X, y = digits.data, digits.target
````

````python
from sklearn.linear_model import LogisticRegression
````

````python
X /= 255.
````

````python
import matplotlib.pyplot as plt
plt.imshow(X[0].reshape(8, 8))
````

````python
clf = LogisticRegression()
clf.fit(X, y)
````

````python
new_data = X[100]
````

````python
from joblib import dump, load
````

````python
saved_model = dump(clf, 'logreg.joblib')
````

````python
reloaded_model = load('logreg.joblib')
````

````python
reloaded_model.predict(X[4].reshape(1, 64))
````

````python
reloaded_model.fit(X, y)
saved_model = dump(reloaded_model, 'logreg.joblib')
````
